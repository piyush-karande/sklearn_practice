```python
from __future__ import print_function, division
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-2.*np.pi, 2.*np.pi, 101)
y = np.sin(x)
```

```python
%matplotlib inline
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.xlabel("$x$")
plt.ylabel("$\sin(x), \cos(x)$", size=20)
plt.legend()
```
